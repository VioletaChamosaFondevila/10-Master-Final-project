## UNIÓN DE LOS DATASET DE PRECIOS Y ANALISIS DE SENTIMIENTOS

```python
import pandas as pd
```

Leemos el dataset del historico de precios y los indicadores técnicos

```python
df_prices_TI = pd.read_csv('BTC_DF.csv', index_col='Date')
```

```python
df_prices_TI.index = pd.to_datetime(df_prices_TI.index)
```

```python
df_prices_TI.head()
```

-------------------

Leemos el dataset del analisis de sentimientos

```python
df_sentiment = pd.read_csv('BTC_sentiment.csv', index_col='Date')
```

```python
df_sentiment.index = pd.to_datetime(df_sentiment.index)
```

```python
df_sentiment = df_sentiment.drop(['articles' ], axis=1)
```

```python
df_sentiment.head()
```

-------------------

Juntamos los datasets por el indice timestamp

```python
df = pd.merge(df_prices_TI, df_sentiment, left_index=True, right_index=True)
```

```python
df
```

Comprobamos si faltan observaciones en el horizonte temporal

```python
pd.date_range(start = '2018-02-25', end = '2023-02-24' ).difference(df.index)
```

Comprobamos los tipos de datos

```python
df.dtypes
```

Vemos si hay algun valor nulo

```python
df.isna().sum()      # Comprobamos que no haya valores nulos
```

Guardamos en dataset final para trabajar con él

```python
df.to_csv('BTC_sentiment_price.csv', encoding='utf-8')
```
